# VR & Age Interaction on Improvement in Learning Outcomes

```python
#Make a boxplot of Age and Improvement_in_Learning_Outcomes
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(x='Age', y='Improvement_in_Learning_Outcomes', data=df, palette='pastel')
plt.title('Improvement in Learning Outcomes by Age', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Improvement', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.9)
plt.show()

#boxplot below seems to confirm even distribution between 0 (no improvement) or 1 (improvement)

#in general, any box plot I made in combinations below proved this point, so I removed them
```

I will group by age below in order to determine if there is a relationship between age, VR use, and effectiveness in learning outcomes. Particularly in adolescent years, there can be major swings in student growth and perspectives due to hormones, peer influence, and maturity, which is why it seems effective to break down by year instead of in buckets of years.

On the other hand, students do not mature at the same rate, so I could also understand the strategy of breaking down ages into buckets (e.g. 12-14, 15-17, 19-21, 23-25, 26-30). However, I will stick with the current strategy.

```python
#group by age, find the mean of improvement in learning outcomes, see above text for rationale

#since the improvement in learning outcomes is simply 1/0, a mean above .5 will let us know if there is more improvement or not
age_groups = df.groupby('Age')
average_improvement = age_groups['Improvement_in_Learning_Outcomes'].mean()
print(average_improvement)
```

Before proceding with age-related considerations, I will investigate how many students, by age, actually use VR.

```python
age_vr_groups = df.groupby(['Age', 'Usage_of_VR_in_Education'])

average_improvement = age_vr_groups['Improvement_in_Learning_Outcomes'].mean()

print(average_improvement)
```

It appears that about half of each age uses VR, enough to proceed with reasonable accuracy.

```python
#visualize improvements by age
average_improvement.plot(kind='bar', color='skyblue')
plt.title('VR Improvement in Learning Outcomes by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Improvement')
plt.xticks(rotation=0)

#label the threshold .5 - anything below this means that VR did not help or even harmed learning
plt.axhline(y=0.5, color='red', linestyle='--', label='Threshold (0.5)')
plt.legend(fontsize='x-small')

plt.show()
```

```python
#pull a list of ages with more improvement than not in order to clarify where investments should be directed

#choose .52 rather than .5 for threshold to be slightly above a draw between 1/0

ages_above_threshold = average_improvement[average_improvement > 0.52].index.tolist()
print("Ages with improvement above 0.5:", ages_above_threshold)
```

```python
#pull a list of ages with less improvement than more because of VR in order to verify where NOT to direct investments
ages_below_threshold = average_improvement[average_improvement < 0.52].index.tolist()
print("Ages with improvement below 0.5:", ages_below_threshold)
```

```python
#visualize with line plot for investor ease of understanding
average_improvement.plot(kind='line', marker='o', color='orange')
plt.title('Average Improvement in Learning Outcomes by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Improvement')
plt.grid(True)

# Add a dashed line at y = 0.52
plt.axhline(y=0.52, color='red', linestyle='--', label='Threshold (0.52)')
plt.legend()  # Show the legend for the threshold line

plt.show()
```

# Transition to interaction between Age/Subject/Region & VR Use on Learning Improvements

Although none of the previous variables showed noticeable impact on learning, the interaction between age, subject, and region with VR use may show an impact on learning improvements. Therefore, group data by age, subject, and region.

```python
# Group data by Age, Subject, and Region and calculate the mean improvement
average_asr_improvement = df.groupby(['Age', 'Subject', 'Region'])['Improvement_in_Learning_Outcomes'].mean()

# Reshape the data for the heatmap
heatmap_data = average_asr_improvement.unstack(level=['Subject', 'Region'])  # Unstack Subject and Region levels
plt.figure(figsize=(20, 11))
# Create the heatmap
sns.heatmap(heatmap_data, annot=True, cmap="YlGnBu")
plt.title("Interaction between Age, Subject, and Region on Learning Outcomes")
plt.xlabel("Subject, Region")
plt.ylabel("Age")
plt.show()
```

```python
# Looking up some other ways to visualize multiple variables, I came across this 3D scatter plot.
#You can zoom in and out. You can also hover over a section of the plot and data shows up in a little text box.
# While it's fun, it's not really easier to see.
#The heatmap above is easier to read, I think.

import plotly.express as px

# Convert categorical variables to numerical for plotting
subject_mapping = {subject: i for i, subject in enumerate(df['Subject'].unique())}
region_mapping = {region: i for i, region in enumerate(df['Region'].unique())}

# Create the scatter plot
fig = px.scatter_3d(
    average_asr_improvement.reset_index(),
    x='Age',
    y='Subject',
    z='Region',
    color='Improvement_in_Learning_Outcomes',
    size='Improvement_in_Learning_Outcomes',
    hover_data=['Age', 'Subject', 'Region', 'Improvement_in_Learning_Outcomes'],
    category_orders={'Subject': list(subject_mapping.keys()),
                    'Region': list(region_mapping.keys())}
)
fig.update_layout(title='Interaction between Age, Subject, and Region on Learning Outcomes')
fig.show()
```

```python
# Looking for other ways to visualize the Age/Subject/Region Interaction and perhaps spot insights
# Create a grid of charts for each subject
g = sns.FacetGrid(average_asr_improvement.reset_index(), col='Subject', row='Region', hue='Region', sharey=False)
g.map(sns.lineplot, 'Age', 'Improvement_in_Learning_Outcomes')
g.add_legend()
plt.show()
```

# Circle Back: Transition to Stress Levels with VR Use

One of the initial research questions concerned finding any associations with stress levels and VR use, and whether this stress level was prohibitive in using VR in education (in this synthetic dataset). It appears that VR has more low-medium stress than high stress, though about a third of students using VR experience high stress levels. Although the data is synthetic, this is one variable where the data could most realistically reflect real-life. Some people could genuinely feel very stressed by VR use because of motion sickness or anxiety, while others would feel little to medium stress.
